# Reaction Time Analysis

In this study we want to measure how strongly information processing will affect a human's reaction time. The Test was split into two parts. For the first one, the participant will only have to react to a visual signal by pressing a button on their keyboard. The screen showed arrow to the left, and the participant then had to press the left arrow on their keyboard as fast as possible. This was repeated 10 times. The process of the test was explained to the participant beforehand, so they knew which button they had to press. The setup for the second test is very similar to the first one, with the only difference being, that the screen will show either a left or a right arrow. The participants then had to press the respective key on their keyboard as fast as possible. This was also be repeated 10 times and was explained before the Test started.
The experiment was planned as a within-subjects design so each participant did both tests.

## Import Packages

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
from scipy import stats
```

## Import CSV Data

```python
reaction_time_results = pd.read_csv('data.csv')
```

## Scatterplot

#### Condition "A". 

```python
#accuracy as x-axes & reaction time as y-axes
x1 = reaction_time_results['correct_key'][reaction_time_results['condition'] == "A"]
y1 = reaction_time_results['reaction_time_s'][reaction_time_results['condition'] == "A"]
```

```python
plt.scatter(x1,y1)

plt.title('Condition A: reaction time per correct key')
plt.xlabel('correct key')
plt.ylabel('reaction time in seconds')
plt.xticks([0,1],['False', 'True'])
plt.xlim(-1,2)
plt.savefig('scatterplot_condition_a.pdf')
```

#### Condition "B"

```python
#accuracy as x-axes & reaction time as y-axes
x2 = reaction_time_results['correct_key'][reaction_time_results['condition'] == "B"]
y2 = reaction_time_results['reaction_time_s'][reaction_time_results['condition'] == "B"]
```

```python
plt.scatter(x2,y2)
plt.xlabel('correct key')
plt.ylabel('reaction time in seconds')
plt.xticks([0,1],['False', 'True'])
plt.xlim(-1,2)
plt.savefig('scatterplot_condition_b.pdf')
```

#### Both Conditions

```python
plt.scatter(x1,y1, label='condition A')
plt.scatter(x2,y2, label='condition B')
plt.legend(loc="upper left")
plt.title('Condition A and B: reaction time ')
plt.xlabel('correct key')
plt.ylabel('reaction time in seconds')
plt.xticks([0,1],['False', 'True'])
plt.xlim(-1,2)
plt.savefig('scatterplot_conditition_a_b.pdf')
```

## Boxplot

```python
plt.boxplot([y1, y2])
plt.title('reaction time per condition')
plt.xlabel('condition')
plt.ylabel('reaction time in sec')
plt.xticks([1,2],['A', 'B'])
plt.savefig('boxplot_reaction_time.pdf')
```

## T-Test

T-Test for "Reaction Time"

```python
stats.ttest_rel(y1,y2)
```

## Means and Standard Deviaton

```python
print( "(M = {} SD = {})".format(np.mean(y1), np.std(y1)) )
print( "(M = {} SD = {})".format(np.mean(y2), np.std(y2)) )
```

## Discussion

The t-test showed that the reaction times to a stimulus with (M = 0.444 SD = 0.106) and without (M = 0.401 SD = 0.099) a mental demand  (M = 16.2, SD = 2.7), are significantly different (t(29) = -2.23, p = .029). We can therefore disprove our H0 as adding a choice significantly changes the reaction time to a stimulus by increasing it.
